##Sentiment Analysis using BERT and Transformers

**Sentiment analysis** (or opinion mining) is a natural language processing (NLP) technique used to determine whether data is positive, negative or neutral. Sentiment analysis is often performed on textual data to help businesses monitor brand and product sentiment in customer feedback, and understand customer needs.

**BERT** stands for Bidirectional Encoder Representations from Transformers. It is pre-trained on a large corpus of unlabelled text including the entire Wikipedia(that’s 2,500 million words!) and Book Corpus (800 million words). The most impressive aspect of BERT is that we  can fine-tune it by adding just a couple of additional output layers to create state-of-the-art models for a variety of NLP tasks.

### Preprocessing

```python
from transformers import AutoTokenizer
checkpoint = "bert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
#example usage

inputs = tokenizer("this is a sentence")
inputs
```

```python
#example usage

tokenizer.convert_ids_to_tokens(inputs["input_ids"])
```

```python
def tokenize_func(sample):
  return tokenizer(sample["sentence"],truncation=True)
```

```python
tokenized_dataset = dataset.map(tokenize_func,batched=True)
tokenized_dataset
```

### Dynamic Padding

```python
from transformers import DataCollatorWithPadding
data_col = DataCollatorWithPadding(tokenizer = tokenizer, return_tensors="tf")
```

```python
#example usage

sample = tokenized_dataset["train"][:10]
sample = {k:v for k,v in sample.items() if k not in ["idx","sentence"]}
[len(x) for x in sample["input_ids"]]
```

```python
#example usage

batch = data_col(sample)
{k:v.shape for k,v in batch.items()}
```

### Fine tuning

```python
from huggingface_hub import HfFolder
import tensorflow as tf
```

```python
train_batch_size = 16
eval_batch_size = 8
num_epochs = 5
init_learning_rate = 5e-5
end_learning_rate = 0.0
hub_token = HfFolder.get_token()
hub_model_id = "sentiment-analysis-sst2"
output_dir = "bert-output"
```

```python
tf_train_dataset = tokenized_dataset["train"].to_tf_dataset(
    columns = ["attention_mask","input_ids","token_type_ids"],
    label_cols = ["labels"],
    shuffle = True,
    collate_fn = data_col,
    batch_size = train_batch_size,
)
```

```python
tf_val_dataset = tokenized_dataset["validation"].to_tf_dataset(
    columns = ["attention_mask","input_ids","token_type_ids"],
    label_cols = ["labels"],
    shuffle = False,
    collate_fn = data_col,
    batch_size = eval_batch_size,
)
```

```python
from tensorflow.keras.optimizers.schedules import PolynomialDecay

num_train_steps = len(tf_train_dataset)*num_epochs

lr_scheduler = PolynomialDecay(
    initial_learning_rate= 5e-5,
    end_learning_rate = 0.0,
    decay_steps = num_train_steps
)

from tensorflow.keras.optimizers import Adam

opt = Adam(learning_rate = lr_scheduler)
```

```python
from transformers import TFAutoModel
model = TFAutoModel.from_pretrained(checkpoint,num_labels = 2)
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
loss=SparseCategoricalCrossentropy(from_logits=True)
model.compile(optimizer=opt,loss = loss)
```

### callbacks

```python
import os
from transformers.keras_callbacks import PushToHubCallback
from tensorflow.keras.callbacks import TensorBoard as TensorBoardCallback
```

```python
callback = []

callback.append(TensorBoardCallback(log_dir = os.path.join(output_dir,"logs")))

if hub_token:
  callback.append(PushToHubCallback(output_dir=output_dir,
                                    tokenizer=tokenizer,
                                    hub_model_id=hub_model_id,
                                    hub_token=hub_token))
```

### training

```python
model.fit(tf_train_dataset, 
          validation_data = tf_val_dataset, 
          callbacks = callback, 
          epochs=5)
```
